```python
import matplotlib.pyplot as plt


workers = [1, 2, 5]
runtime = [216.4, 117.8, 88.1]

plt.figure(figsize=(8, 5))
plt.plot(workers, runtime, marker='o', linestyle='-', label='Runtime (s)')
plt.xlabel('Number of Workers')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime vs Number of Workers')
plt.xticks(workers)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
